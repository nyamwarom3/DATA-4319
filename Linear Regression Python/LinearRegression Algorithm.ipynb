## LINEAR REGRESSION ALGORITHM IN PYTHON

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
#Reading data 
data = pd.read_csv("test.csv")
data.head()
```

```python
X = data['x'].values
Y = data['y'].values

#Mean of X and Y
mean_x = np.mean(X)
mean_y = np.mean(Y)

#Total number of values
m = len(X)

#formula to calculate B1(slope) and B0(intercept)
numer = 0
denom = 0

for i in range(m):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2 
    
B1 = numer/denom
B0 = mean_y - (B1*mean_x)

print(B1,B0)
```

```python
#Plotting values and regression line 

max_x = np.max(X) + 100
min_x = np.min(X) - 100

#Calculating line values x and y
x = np.linspace(min_x, max_x, 1000)
y = B0 + B1 * x

#Plotting line
plt.plot(x, y, color = '#58b970', label = 'Regression Line')
#Plotting scatter points
plt.scatter(X, Y, c = '#ef5423', label = 'Scatter Plot')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = X.reshape((m, 1))

#Creating model
reg = LinearRegression()

#Fitting training data
reg = reg.fit(X, Y)

#Y prediction
Y_pred = reg.predict(X)

#Calculating R2 score
r2_score = reg.score(X, Y)
print(r2_score)
```

**From calculating the r-squared we get a value of 98.9%, showing that the model is very accurate. In other words, an r-squared of 98.9% reveals that 98.9% of the data fit the regression model.**
